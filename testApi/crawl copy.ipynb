```python
import requests 
from bs4 import BeautifulSoup
import json
```

```python
url = 'https://4menshop.com/ao-polo-rugby-shirt-nam.html'
response = requests.get(url)
```

```python
response.status_code
```

```python
# Kiểm tra nếu yêu cầu thành công(mã trạng thái 200)
if response.status_code == 200:
    #phân tích cú pháp HTML của trang web
    soup = BeautifulSoup(response.content, 'html.parser')  

    table = soup.find_all('div', class_='product-info')

    data =[]
    for roa in table:
        name = roa.find('a', class_='product-title').text.strip()
        
        price = roa.find('span', class_='product-price').text.strip()
        
        words = name.split()
        product_code = words[-1]
        type_ = words[1]

        data.append({
            "id": product_code, 
            "name": name, 
            "type": type_, 
            "price": price})
        print("id: ",product_code, "\nname: ", name,"\ntype: ",type_,"\nprice: ", price, "\n\n")
    #xuat du lieu ra file Json
    with open("output.json", "w", encoding="utf-8") as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

    print("Dữ liệu đã được ghi ra file output.json")
```
